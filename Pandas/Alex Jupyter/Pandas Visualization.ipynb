## Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df1 = pd.read_csv(r"C:\Users\bhanu\iCloudDrive\Documents\Alex the Analyst Course\Pandas\Data\Ice Cream Ratings.csv")
df1 = df1.set_index('Date')
```

```python
df1.plot()
```

```python
df1.plot(kind = 'line', title = 'Ice Cream Ratings')
```

```python
df1.plot(kind = 'bar', stacked = True, xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df1.plot.barh(stacked = True)
# or df1.plot(kind = 'barh', stacked = True)
```

```python
df1['Flavor Rating'].plot(kind = 'bar', stacked = True, xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df1[['Flavor Rating', 'Overall Rating']].plot(kind = 'barh', stacked = True, xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df1.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s = 75, c = 'Turquoise')
```

```python
df1.plot.hist()
```

```python
df1.plot.hist(bins = 20)
```

```python
df1.boxplot()
```

```python
df1.plot.area()
```

```python
df1.plot.area(figsize = (10,4))
```

```python
df1.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```

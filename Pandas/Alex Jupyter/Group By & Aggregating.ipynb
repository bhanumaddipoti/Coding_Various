# Group By & Aggregating

```python
import pandas as pd
```

```python
df1 = pd.read_csv (r"C:\Users\bhanu\iCloudDrive\Documents\Alex the Analyst Course\Pandas\Data\Flavors.csv")
```

```python
groupby_base_flavor = df1.groupby('Base Flavor')
```

```python
groupby_base_flavor_and_Liked = df1.groupby(['Base Flavor', 'Liked'])
```

```python
groupby_base_flavor.mean(numeric_only = True)
```

```python
df1.groupby('Base Flavor').mean(numeric_only = True)
```

```python
groupby_base_flavor.count()
```

```python
groupby_base_flavor.min()
```

```python
groupby_base_flavor.max()
```

```python
groupby_base_flavor.sum()
```

```python
groupby_base_flavor.sum(numeric_only = True)
```

```python
groupby_base_flavor.agg({'Flavor Rating': ['mean', 'max', 'count', 'sum']})
```

```python
groupby_base_flavor.agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
groupby_base_flavor.agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']}, numeric_only = True)
```

```python
groupby_base_flavor_and_Liked.agg({'Flavor Rating': ['mean', 'max', 'count', 'sum'], 'Texture Rating': ['mean', 'max', 'count', 'sum']})
```

```python
groupby_base_flavor_and_Liked.describe()
```

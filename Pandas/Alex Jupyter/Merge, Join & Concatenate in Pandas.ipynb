# Merge, Join & Concatenate in Pandas


```python
import pandas as pd
```

```python
df1 = pd.read_csv (r"C:\Users\bhanu\iCloudDrive\Documents\Alex the Analyst Course\Pandas\Data\LOTR.csv")
```

```python
df2 = pd.read_csv (r"C:\Users\bhanu\iCloudDrive\Documents\Alex the Analyst Course\Pandas\Data\LOTR 2.csv")
```

## Merge

```python
df1.merge(df2)
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left', on = 'FellowshipID')
```

```python
df1.merge(df2, how = 'right', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'cross')
```

```python
df1.merge(df2, how = 'cross', suffixes = ('_df1', '_df2'))
```

## Join

```python
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
```

```python
df3 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_left', rsuffix = '_right', how = 'outer')
```

```python
df3
```

```python
df4 = df1.set_index(['FellowshipID','FirstName']).join(df2.set_index(['FellowshipID','FirstName']), lsuffix = '_left', rsuffix = '_right', how = 'outer')
```

```python
df4
```

## Concatenate

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'outer')
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1, ignore_index=True) 
# Append no longer Works in PD Pandas
```
